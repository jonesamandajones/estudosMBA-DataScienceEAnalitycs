## <font color=purple>**Assimetria e Curtose:**</font>

### **Assimetria**

>* Curva simétrica (normal) => Média = Mediana = Moda (x̅ = Md = Mo)
>
>* Curva assimétrica à direita (cauda mais longa à dir.) => Média > Mediana (x̅ > Md)
>
>* Curva assimétrica à esquerda (cauda mais longa à esq.) => Média < Mediana (x̅ < Md)
>
>
>
> ### **Coeficiente de assimetria de Fischer:**
>
>* No excel => **DISTORÇÃO(num1; num2)**
>
>
>
>* $g¹ = 0$ => indica curva simétrica
>
>* $g¹ > 0$ => indica curva assimétrica positiva (à direita)
>
>* $g¹ < 0$ => indica curva assimétrica negativa(à esquerda)


### **Curtose**

>#### Análise do achatamento da curva de distribuição (se a curva apresenta menor ou maior dispersão em torno da média)
>
>* Curva Mesocúrtica (normal)
>
>* Curva Platicurtica (achatada)
>
>* Curva Leptocúrtica (alongada)
>
>
>
>#### Coeficiente de Curtose de Fischer:
>
>* No Excel => **CURT(num1;num2)**
>
>
>
>* g2 = 0 => indica curva com distribuição normal
>
>* g² > 0 => indica curva com distribuição alongada
>
>* g² < 0 => indica curva com distribuição achatada



# <font color=purple>**Relacionando variáveis: Medidas de análises bivariadas**</font>

---

> Variáveis **qualitativas**: Análise da associação por teste **qui-quadrado**($X²$)
>
> Variáveis **quantitativas**: Análise da correlação por meio da **covariância** e **correlação** de Pearson

---

## <font color=purple>**Teste de associação qui-quadrado: variáveis ***qualitativas*****</font>

### **Frequências Absolutas ***observadas*****
>
> 
>
> * Tabela de distribuição conjunta de frequências: **Tabela de contigência** ou Tabela de classificação cruzada
>
> 
>
> * Cálculo: Soma da *frequência absoluta* das variáveis que coincidem em cada categoria


### **Frequências Absolutas ***esperadas*****
>
> 
>
> * Cálculo: total da linha * total da coluna / total da amostra
>
> 


### ***Resíduos***
>
> 
>
> Cálculo: Frequência absoluta observada - Frequência absoluta esperada
>
> 


### **Cálculo dos valores qui-quadrado ($X²$):** 
>
>
>
> Cálculo: Resíduo ^ 2 / Frequência absoluta esperada
>
>
>


### **Valores do teste qui-quadrado**
>
> **Qui-quadrado total ou $X²$ total**: somar todos os valores de $X²$ individuais da tabela qui-quadrado
>
>**Grau de liberdade**: (número de linhas - 1) x (número de colunas - 1) ou $(l-1).(j-1)$
>
>**Nível de significância**: Varia entre 1% e 10%, mais utilizado como 5%.
>
>**Valor crítico**: No excel => **INV.QUIQUA.CD(nível de significância;grau de liberdade)**
>
>**p-valor**: Cálculo da área à direita do valor $X²$ total. No excel existem duas formas de calcular:
> 
> * **DIST.QUIQUA.CD($X²$ total; grau de liberdade)**
>
> * **TESTE.QUIQUA(total de ***frequências absolutas observadas***; total de ***frequencias absolutas esperadas***)**


### **Testando o qui-quadrado**
>
> Resultados do teste qui-quadrado: **hipótese nula (H°)** ou **hipótese positiva (H¹)**
>
> * H°: Variáveis se associam aleatoriamente
>
> * H¹: Variáveis se associam de forma não aleatória
>
> #### Teste a partir do *valor crítico*:
>
> Se $X²$ total < valor crítico => H° (As variáveis não se relacionam)
>
> Se $X²$ total > valor crítico => H¹ (As variáveis se relacionam dentro do grau de significância escolhido)
>
> #### Teste a partir do *p-valor*:
>
> Se **Nível de significância** > p-valor => H° (As variáveis não se relacionam)
>
> Se **Nível de significância** < p-valor => H¹ (As variáveis se relacionam dentro do grau de significância escolhido)
>

## <font color=purple> **Coeficiente de correlação de Pearson: ***variáveis quantitativas*****</font>

>
>
>
> Calcula-se a **covariância $cov(X,Y)$** para se obter o **coeficiente de correlação de Pearson $Pxy$**
>
> * $Pxy$ = -1 => correlação **perfeita negativa**, as variáveis caminham juntas mas em **sentidos opostos** em relação à média
>
> * $Pxy$ = 0 => **não há correlação** entre as variáveis
>
> * $Pxy$ = 1 => correlação **perfeita positiva**, as variáveis caminham juntas no mesmo sentido
>
> #### Covariância no Excel ($cov(X,Y)$) => **COVARIACAO.S(matriz1;matriz2)**
>
> #### Correlação ($Pxy$) no Excel => **CORREL(matriz1;matri2)**






## <font color=purple> ** Distribuições de probabilidade** </font>

### **Característica primodial**
* Variável aleatória discreta => 0, 1, 2, 3 ...
* Variável aleatória contínua => 0.0, 0.1, 0.2, 0.3 ...

### **Distribuições de probabilidade**:
* Uniforme
* Bernoulli
* Binomial 
* Binomial negativa
* Poisson

#### **Distribuições de probabilidades em variáveis discretas:**

#### Distribuição uniforme discreta
>Todas as possibilidades tem a mesma probabilidade de ocorrência (dado, moeda etc.) 
>
>

#### Distribuição de Bernoulli (logística binária)
>Os valores das variáveis assumem apenas dois resultados possíveis, chamados de *sucesso* ($x=1$) ou *fracasso* ($x=0$)
>
>Sendo a probabilidade de sucesso ($p$) ou de fracasso ($p-1$) quando ocorre apenas um experimento.
>
>

#### Distribuição binominal (logística multinomial)
>Quando há ($n$) repetições do experimento de Bernoulli e a probabilidade de sucesso ($p$) é constante em todas as repetições
>
>Este modelo se baseia em encontrar a *probabilidade de sucesso* ($p$) a partir da *quantidade de sucessos* ($k$) e *repetições do experimento* ($n$)
>
>

#### Distribuição binomial negativa
>Este modelo se baseia em encontrar a   *probabilidade de sucesso* ($p$) a partir da *repetições do experimento* ($n$) e *quantidade de sucessos* ($k$).
>
>
