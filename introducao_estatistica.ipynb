# <font color=purple>**Relacionando variveis: Medidas de anlises bivariadas**</font>

---

> Variveis **qualitativas**: Anlise da associao por teste **qui-quadrado**($X$)
>
> Variveis **quantitativas**: Anlise da correlao por meio da **covarincia** e **correlao** de Pearson

---

## <font color=purple>**Teste de associao qui-quadrado: variveis ***qualitativas*****</font>

### **Frequncias Absolutas ***observadas*****
>
> 
>
> * Tabela de distribuio conjunta de frequncias: **Tabela de contigncia** ou Tabela de classificao cruzada
>
> 
>
> * Clculo: Soma da *frequncia absoluta* das variveis que coincidem em cada categoria


### **Frequncias Absolutas ***esperadas*****
>
> 
>
> * Clculo: total da linha * total da coluna / total da amostra
>
> 


### ***Resduos***
>
> 
>
> Clculo: Frequncia absoluta observada - Frequncia absoluta esperada
>
> 


### **Clculo dos valores qui-quadrado ($X$):** 
>
>
>
> Clculo: Resduo ^ 2 / Frequncia absoluta esperada
>
>
>


### **Valores do teste qui-quadrado**
>
> **Qui-quadrado total ou $X$ total**: somar todos os valores de $X$ individuais da tabela qui-quadrado
>
>**Grau de liberdade**: (nmero de linhas - 1) x (nmero de colunas - 1) ou $(l-1).(j-1)$
>
>**Nvel de significncia**: Varia entre 1% e 10%, mais utilizado como 5%.
>
>**Valor crtico**: No excel => **INV.QUIQUA.CD(nvel de significncia;grau de liberdade)**
>
>**p-valor**: Clculo da rea  direita do valor $X$ total. No excel existem duas formas de calcular:
> 
> * **DIST.QUIQUA.CD($X$ total; grau de liberdade)**
>
> * **TESTE.QUIQUA(total de ***frequncias absolutas observadas***; total de ***frequencias absolutas esperadas***)**


### **Testando o qui-quadrado**
>
> Resultados do teste qui-quadrado: **hiptese nula (H)** ou **hiptese positiva (H)**
>
> * H: Variveis se associam aleatoriamente
>
> * H: Variveis se associam de forma no aleatria
>
> #### Teste a partir do *valor crtico*:
>
> Se $X$ total < valor crtico => H (As variveis no se relacionam)
>
> Se $X$ total > valor crtico => H (As variveis se relacionam dentro do grau de significncia escolhido)
>
> #### Teste a partir do *p-valor*:
>
> Se **Nvel de significncia** > p-valor => H (As variveis no se relacionam)
>
> Se **Nvel de significncia** < p-valor => H (As variveis se relacionam dentro do grau de significncia escolhido)
>

## <font color=purple> **Coeficiente de correlao de Pearson: ***variveis quantitativas*****</font>

>
>
>
> Calcula-se a **covarincia $cov(X,Y)$** para se obter o **coeficiente de correlao de Pearson $Pxy$**
>
> * $Pxy$ = -1 => correlao **perfeita negativa**, as variveis caminham juntas mas em **sentidos opostos** em relao  mdia
>
> * $Pxy$ = 0 => **no h correlao** entre as variveis
>
> * $Pxy$ = 1 => correlao **perfeita positiva**, as variveis caminham juntas no mesmo sentido
>
> #### Covarincia no Excel ($cov(X,Y)$) => **COVARIACAO.S(matriz1;matriz2)**
>
> #### Correlao ($Pxy$) no Excel => **CORREL(matriz1;matri2)**






## <font color=purple> ** Distribuies de probabilidade** </font>

### **Caracterstica primodial**
* Varivel aleatria discreta => 0, 1, 2, 3 ...
* Varivel aleatria contnua => 0.0, 0.1, 0.2, 0.3 ...

### **Distribuies de probabilidade**:
* Uniforme
* Bernoulli
* Binomial 
* Binomial negativa
* Poisson

#### **Distribuies de probabilidades em variveis discretas:**

#### Distribuio uniforme discreta
>Todas as possibilidades tem a mesma probabilidade de ocorrncia (dado, moeda etc.) 
>
>

#### Distribuio de Bernoulli (logstica binria)
>Os valores das variveis assumem apenas dois resultados possveis, chamados de *sucesso* ($x=1$) ou *fracasso* ($x=0$)
>
>Sendo a probabilidade de sucesso ($p$) ou de fracasso ($p-1$) quando ocorre apenas um experimento.
>
>

#### Distribuio binominal (logstica multinomial)
>Quando h ($n$) repeties do experimento de Bernoulli e a probabilidade de sucesso ($p$)  constante em todas as repeties
>
>Este modelo se baseia em encontrar a *probabilidade de sucesso* ($p$) a partir da *quantidade de sucessos* ($k$) e *repeties do experimento* ($n$)
>
>

#### Distribuio binomial negativa
>Este modelo se baseia em encontrar a   *probabilidade de sucesso* ($p$) a partir da *repeties do experimento* ($n$) e *quantidade de sucessos* ($k$).
>
>
